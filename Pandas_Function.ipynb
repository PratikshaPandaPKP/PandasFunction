# Pandas Function


SHOW CASE/EXPLAIN PANDAS FUNCTIONS FOR Exploratory Data Analysis

Dataset Link :- https://github.com/jbolla368/Pandas-using-in-EDA/blob/master/housing_dataset.csv

```python
#Import Libraries

import pandas as pd
import numpy as np

#Load Dataset

data = pd.read_csv("F:\housing_dataset.csv")
```

```python
#Head and Tail Function 

data.head() #Shows first 5 rows
```

```python
data.tail() #shows last 5 rows
```

```python
#Dataframe.info()  get a concise summary of the dataframe.

data.info()
```

```python
#Dtypes shows the data type of each column.

data.dtypes
```

```python
#Shape and Size (It shows the number of dimensions as well as the size in each dimension.)

print(data.shape)

print(data.size)
```

```python
#describe() quick statistical summary for every numerical column

data.describe()
```

```python
#Sample (select values randomly from a Series or DataFrame)

data.sample(n=5) #gives any 5 random rows
```

```python
#Identifying Missing Values Isnull()

data.isnull()
```

```python
#Isna (returns a dataframe filled with boolean values with true indicating missing values)

data.isna().any()
```

```python
#Identifying Missing Values Sum() (Show the total number of missing values)

data.isnull().sum()
```

```python
#Nunique (counts the number of unique entries over columns or rows)

data.nunique()
```

```python
#Index() and column() 

print(data.index)

print(data.columns)
```

```python
#Memory_usage() (returns how much memory each column uses in bytes)

data.memory_usage()
```

```python
#nsmallest() and nlargest() (Shows the n number of smallest/largest values for specific coulmn )

data.nsmallest(5,'median_house_value')
```

```python
data.nlargest(5,'median_house_value')
```

```python
#Loc and iloc (loc: select by labels & iloc: select by positions)

data.loc[:5,['median_income','median_house_value']]
```

```python
data.iloc[:4,:6]
```

```python
#Slicing

data[0:4] #slice data from row no 0 till 4
```

```python
#Groupby 

data[['housing_median_age','median_house_value','total_rooms']].groupby(['housing_median_age','median_house_value']).mean()
```

```python
#Sorting (sort_index() method by passing the axis arguments and the order of sorting, DataFrame can be sorted)

data.sort_index(axis=1,ascending=True)
```

```python
#sorting by values

data.sort_values(by='median_house_value')
```

```python
#Dropna() (remove a row or a column from a dataframe which has a NaN or no values in it)

data.dropna()
```

```python
#Query (filter a dataframe based on a condition or apply a mask to get certain values)

data.query('40000 < median_house_value < 100000')[:8]
```

```python
#Insert (add a new column to a dataframe, it is added at the end by default)

new_col = np.random.randn(20640)

data.insert(2,'new_col',new_col)
```
